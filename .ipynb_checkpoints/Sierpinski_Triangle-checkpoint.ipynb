# Fun With Fractals

This is a short notebook to explore both binary trees and their use in constructing the [SierpiÅ„ski triangle](https://en.wikipedia.org/wiki/Sierpi%C5%84ski_triangle). The original inspiration came from [Doodling in Math Class](https://www.youtube.com/watch?v=e4MSN6IImpI&list=PLF7CBA45AEBAD18B8), and more information can be found on [github](https://github.com/brettrhenderson/pyFractals).

```python
import matplotlib.pyplot as plt
import numpy as np
from matplotlib import animation, rc
from IPython.display import HTML
```

## A Couple of Tree-Building Classes

We make basic binary trees, which can be build exclusively from the **Node** class. Each node can have a left and/or right child, and its value is just its location in a 2D cartesian coordinate system. The **Tree** class defines a special type of tree that ends up building a Sierpinski triangle and includes the methods for growing the tree.  The special bit about this type of tree is that its branches are all equal in length and angle, and every time two branches meet at their bottoms, no children grow from them.

```python
class Tree():
    """
    Tree-like object.
    """
    def __init__(self, root, dx, dy, ax=None, **kwargs):
        self.root = root
        self.leaves = [root]
        self.dx = dx
        self.dy = dy
        self.style = kwargs
        self.ax = ax
        #root.draw()
    def add_layer(self, plot=False):
        new_leaves = []
        for leaf in self.leaves:
            leaf.left = Node([leaf.val[0]-self.dx, leaf.val[1]-self.dy], leaf)
            leaf.right = Node([leaf.val[0]+self.dx, leaf.val[1]-self.dy], leaf)
            # draw the new branches
            if plot:
                leaf.left.draw(self.ax, **self.style)
                leaf.right.draw(self.ax, **self.style)
            if len(new_leaves) and leaf.left.val == new_leaves[-1].val:
                new_leaves.pop()
                new_leaves.append(leaf.right)
            else:
                new_leaves += [leaf.left, leaf.right]
        self.leaves = new_leaves
        
    def add_layers(self, n, plot=False):
        if n > 30:
            raise ValueError("Too many layers will really crap out your computer!\n"
                             "Try L <= 30 or disable this error at your own risk.")
        for _ in range(n):
            self.add_layer(plot)


class Node():
    """
    Nodes of a tree
    
    Parents, as well as left and right children are other nodes or None
    Values are tuples or lists of the location of the node in Cartesian coordinates.
    """
    def __init__(self, value, parent=None, left=None, right=None):
        self.par = parent
        self.val = value
        self.left = left
        self.right = right
    def draw(self, ax=None, **kwargs):
        if ax is None:
            ax = plt.gca()
        if self.par is None:
            ax.scatter(self.val[0], self.val[1], **kwargs)
        else:
            ax.plot([self.par.val[0], self.val[0]], [self.par.val[1], self.val[1]], **kwargs)
```

## Tree Traversal Algorithms

Define a few basic tree traversal algorithms. These can be used later when animating by returning the tree nodes in the order in which they are searched, such that they can be animated in that order.  

```python
def dfs(node):
    """
    Traditional depth-first tree search (left branch, then node, then right branch)
    
    Help from tutorial at https://www.geeksforgeeks.org/dfs-traversal-of-a-tree-using-recursion/
    """
    if node is not None:
        yield from dfs(node.left)
        yield node
        yield from dfs(node.right)

def preorder(node):
    """
    Pre-order tree search (node, left branch, then right branch)
    
    Help from tutorial at https://www.geeksforgeeks.org/dfs-traversal-of-a-tree-using-recursion/
    """
    if node is not None:
        yield node
        yield from preorder(node.left)
        yield from preorder(node.right)

def postorder(node):
    """
    Post-order.  Left, right, node
    
    Help from tutorial at https://www.geeksforgeeks.org/dfs-traversal-of-a-tree-using-recursion/
    """
    if node is not None:
        yield from postorder(node.left)
        yield from postorder(node.right)
        yield node
        
def bfs(node):
    """
    Breadth-first search
    
    Help from tutorial at https://www.geeksforgeeks.org/level-order-tree-traversal/
    """
    if node is None:
        return
    queue = [node]
    while len(queue) > 0:
        yield queue[0]
        temp = queue.pop(0)
        
        if temp.left is not None:
            queue.append(temp.left)
        if temp.right is not None:
            queue.append(temp.right)
            
def ubfs(node):
    """Upside-down breadth-first search"""
    nodes = list(bfs(node))
    nodes.reverse()
    for node in nodes:
        yield node
```

## Animation 

Here are the functions for animating the construction of the special **Tree** above. The animation proceeds one node at a time, in the order specified by the selected tree traversal algorithm. The branches attached to each node are inimated by swinging down from a vertical position into their final states.

```python
def frames(nodes, n):
    """
    Generator of frames for animation.
    
    i is the node, j is the animation step for that node.
    """
    for i, node in enumerate(nodes):
        for j in range(n):
            yield node, i, j

def animate_sierpinsky(tree, order, N, interval=20, save_count=10**4):
    """
    Animate drawing the triangle using the ordering order.
    
    Order is a generator returning the nodes in the order you want.
    """
    
    nodes = order(tree.root)

    # the sweep for each new edge
    r = np.sqrt(dx**2+dy**2)
    final = np.pi + np.arctan(dy/dx)

    # First set up the figure, the axis, and the plot element we want to animate
    fig = plt.figure()
    ax = fig.add_subplot(111, autoscale_on=False, aspect='equal')
    line,  = ax.plot([], [], lw=2)
    lines = [line]
    dthetas = np.linspace(0, np.sqrt(final-np.pi/2), N)**2
    n = 0

    # initialization function: plot the background of each frame
    def init():
        lines[-1].set_data([], [])
        return lines

    # animation function.  This is called sequentially
    def animate(nodeij):
        global thetas
        node, i, j = nodeij
        if node.par is None:  # root
            return lines
        if j == 0:    # new node
            if i == 0:
                #first node, set axis lims
                ax.set_xlim(min(node.val[0], node.par.val[0])-dx, max(node.val[0], node.par.val[0])+dx)
                ax.set_ylim(min(node.val[1], node.par.val[1])-dy, max(node.val[1], node.par.val[1])+dy)
            # rescale xlim/ylim as needed
            xlim = ax.get_xlim()
            ylim = ax.get_ylim()
            if node.val[0] > xlim[1]:
                ax.set_xlim(xlim[0], node.val[0]+dx)
            elif node.val[0] < xlim[0]:
                ax.set_xlim(node.val[0]-dx, xlim[1])
            if node.val[1] < ylim[0]:
                ax.set_ylim(node.val[1]-dy, ylim[1])
            elif node.par.val[1] > ylim[1]:
                ax.set_ylim(ylim[0], node.par.val[1]+dy)

            # is it a right or left:
            if node.val[0] > node.par.val[0]:  # right
                thetas = np.pi/2 - dthetas
            else:
                thetas = np.pi/2 + dthetas

            new_line, = ax.plot([], [], lw=2)
            lines.append(new_line)
        theta = thetas[j]
        lines[-1].set_data([node.par.val[0], r*np.cos(theta)+node.par.val[0]], 
                           [node.par.val[1], r*np.sin(theta)+node.par.val[1]])
        return lines
    plt.axis('off')
    # call the animator.  blit=True means only re-draw the parts that have changed.
    return animation.FuncAnimation(fig, animate, init_func=init, repeat=False, save_count=save_count,
                                   frames=frames(nodes, N), interval=interval, blit=True)
```

### Set up an Animation

The top several variable definitions allow you to control the look of the animation.  Keep **N** >= 1 since it is the number of steps that are drawn when swinging a branch down to final position during the animation. **L** lets you control tree size, but I'd recommend keeping it below 35, or else your computer may run out of memory. In addition, **o** lets you choose the order you'd like to animate by specifying one of the tree traversal algorithms defined above. You can also adjust the speed of the animation by changing the **interval** between frames.

```python
# Run it
N=3  # number of animation steps per branch
L = 15  # number of layers of tree to animate
o = postorder  # order to draw the tree -- dfs, bfs, ubfs, preorder, postorder
interval = 20  # ms, the time between animation frames

dx, dy = (1, 1)  # the length/width of your branches
tree = Tree(Node([0,0]), dx, dy)  # initiate a tree with an origin at 0,0
tree.add_layers(L)  # grow the tree
anim = animate_sierpinsky(tree, o, N, interval=interval, save_count=10**4)  # animate the drawing
```

## Building a Tree

Here is a simple way to build a Sierpinski triangle without animating.  Just set up your axes and then add the `plot=True` argument to the `tree.add_layers()` method.

```python
fig = plt.figure(figsize=(6,6))
ax = fig.add_subplot(111)
tree = Tree(Node([0,0]), 1, 1, ax=ax)
tree.add_layers(3, plot=True)
plt.axis('off')
plt.axis('equal')
plt.show()
```
